```python
from prophet import Prophet
```

```python
h=pd.DataFrame({'holiday':['new year','Independence day'],
                'ds':[pd.to_datetime('2020-01-01'),pd.to_datetime('2020-08-15')]})
h
```

```python
model_prophet=Prophet()#h=holidays
```

```python
#training
model_prophet.fit(df_india)
```

```python
future=model_prophet.make_future_dataframe(periods=30,freq='D')
future
```

```python
#forecast
forecast=model_prophet.predict(future)
forecast
#yhat= final output(yhat_lower	yhat_upper)
#trend-trend(trend_lower	trend_upper)
#weekly-seasonality(weekly_lower	weekly_upper)
#multiplicative_terms-holiday(ultiplicative_terms_lower	multiplicative_terms_upper)
#additive_terms-error(additive_terms_lower	additive_terms_upper)
```

```python
final_forecast=forecast[['ds','yhat','yhat_lower','yhat_upper']]
final_forecast
```

```python
#plotting
model_prophet.plot(final_forecast)
```
